```python
def cosine_similarity(v1, v2):
    v1_norm = tf.nn.l2_normalize(v1, axis=-1)
    v2_norm = tf.nn.l2_normalize(v2, axis=-1)
    return tf.reduce_sum(tf.multiply(v1_norm, v2_norm), axis=-1)
```

```python
def enhanced_similarity(features, user_input, ratings, sentiments, alpha=0.4, beta=0.6):
    # Calculate cosine similarity between user input and feature vectors
    sim_scores = cosine_similarity(features, user_input)
    ratings_scaled = (ratings - tf.reduce_min(ratings)) / (tf.reduce_max(ratings) - tf.reduce_min(ratings))
    sentiments_scaled = (sentiments - tf.reduce_min(sentiments)) / (tf.reduce_max(sentiments) - tf.reduce_min(sentiments))
    final_scores = alpha * sim_scores + beta * (0.4 * ratings_scaled + 0.6 * sentiments_scaled)
    return final_scores
```

```python
features = tf.constant([sublist[0:-2] for sublist in np_data], tf.float32)
# Bahari, Budaya, Cagar Alam, Pusat Perbelanjaan, Tempat Ibadah, Taman Hiburan
user_input = tf.constant([0, 1, 0, 1, 1, 0, 0, 1], tf.float32) # budaya,cagar alam, pusat perbelanjaan, kemarau
ratings = tf.constant([row[-1] for row in np_data],  tf.float32)
sentiments = tf.constant([row[-2] for row in np_data], tf.float32)
```

```python
print(features)
```

```python
tf.shape(features)
```

```python
recommendations = enhanced_similarity(features, user_input, ratings, sentiments)
```

```python
print(recommendations)
```

```python
sorted_indices = tf.argsort(recommendations, direction='DESCENDING')
```

```python
N = 10  # Number of recommendations you want
top_N_indices = sorted_indices[:N]

# Print the indices of the top N recommendations
print("Top N recommendation indices:", top_N_indices)
```

```python
top_N_df = pd.DataFrame({'Place_Id': top_N_indices.numpy()})
recommended_places = pd.merge(top_N_df+1, df_tourism, on='Place_Id', how='inner')
```
